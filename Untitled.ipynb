```python
import tensorflow as tf
from tensorflowjs import converters
import numpy as np
np.bool_ = np.dtype('bool')

# Temporarily restore the deprecated np.object for compatibility
np.object = np.dtype('object')

# Load the pre-trained model from the .h5 file
model = tf.keras.models.load_model('NumOcr_model_weight.h5')

# Convert and save the model to JSON format
converters.save_keras_model(model, 'NumOcr_model_weight')

# Remove the temporary modification to np.object
del np.object

print("Conversion completed.")
```
